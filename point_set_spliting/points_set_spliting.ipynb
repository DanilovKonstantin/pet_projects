### Splitting set of points to latitudinal profiles
it is assumed that all points in the profile deviate from the initial latitude by less than 30 m

```python
#import libraries
import pandas as pd
from datetime import datetime, timedelta
from geopy.distance import geodesic as GD
```

```python
#choose only data with information about points

df_clean=df.loc[0:max_ind,['ID','name','lat','lon', 'ele', 'desc','CreationTime']]
df_clean.head()
```

```python
#convert time to DateTime format
df_clean['CreationTime'] = pd.to_datetime(df_clean['CreationTime'])
#convert to float 
df_clean[['lat','lon','ele']] = df_clean[['lat','lon','ele']].astype(float)
df_clean.info()
```

```python
#clean desc
df_clean.head()
```

```python
#latitude distance calculation function
def diastance_lat(row):
    # empty lists for coordinates
    cordx = []
    #list of coordinates of the calculated point with longitude equal to the reference
    cordx.append(row['lat'])
    cordx.append(cord0[1])
    #calculate the distance by latitude in m
    d=GD(cord0,cordx).m
    return round(d)
    
```

```python
#distance calculation function
def diastance1(row):
    # empty lists for coordinates
    cordx = []
    #list of reference point coordinates 
    cordx.append(row['lat'])
    cordx.append(row['lon'])
    #calculate the distance in m
    d=GD(cord0,cordx).m
    return round(d)
```

```python
# create list of first points in profiles
start_points = ['11_1', '10_1','9_1','8_1','7_1','6_1','5_1','4_1','3_1','2_1','1_1']
i=0
for point in start_points:
    i=i+1
    pr_number='pr_'+str(i)
    first_ind = df_clean[df_clean['name'] == point].index.tolist()[0]
    #create a list of coordinates of the zero point
    cord0 = []
    cord0.append(df_clean.loc[first_ind,'lat'])
    cord0.append(df_clean.loc[first_ind,'lon'])
    #distance on lat calculation 
    df_clean['distance_lat']=df_clean.apply(diastance_lat, axis=1)
    #save points nearst to zero point in latitude
    pr = df_clean.query('distance_lat<30')
    # distance to zero point
    pr['distance'] = pr.apply(diastance1, axis=1)
    pr= pr.sort_values(by = 'distance').reset_index(drop=True)
    #rename to correct point name
    pr['name'] = pr.index.tolist()
    pr['name'] = pr_number+'_'+pr['name'].astype('str')
    pr=pr[['ID','name', 'lat','lon','ele','desc', 'CreationTime']]
    #save profil to csv
    pr.to_csv('profile_'+pr_number+'.csv', index= False )

    
```
